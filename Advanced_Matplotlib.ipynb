### 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
x
```

```python
xx, yy = np.meshgrid(x,y)
xx.shape
```

```python
xx, yy = np.meshgrid(x,y)
yy.shape
```

```python
z = xx*2 + yy*2
z.shape
```

```python
fig = plt.figure(figsize=(18,8))
ax = plt.subplot(projection = '3d')
ax.plot_surface(xx,yy,z, cmap='coolwarm', edgecolor='none', linewidth=0.2)
plt.show()
```

```python
z = xx**2 + yy**2
z.shape

fig = plt.figure(figsize=(18,8))
ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z, cmap='coolwarm', edgecolor='none', linewidth=0.2)
fig.colorbar(p)
plt.show()
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(18,8))
ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z, cmap='coolwarm', edgecolor='none', linewidth=0.2)
fig.colorbar(p)
plt.show()
```

### Contour Graph

```python
# Converts 3D graph into 2D graph
```

```python
z = xx**2 + yy**2
z.shape

fig = plt.figure(figsize=(18,8))
ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z, cmap='coolwarm', edgecolor='none', linewidth=0.2)
fig.colorbar(p)
plt.show()
```

```python
# 2D graph of above 3D graph - contour

fig = plt.figure(figsize=(18,8))
ax = plt.subplot()
p = ax.contour(xx,yy,z, cmap='coolwarm')
fig.colorbar(p)
plt.show()
```

```python
# contourf (filled 2D graph)

fig = plt.figure(figsize=(18,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z, cmap='coolwarm')
fig.colorbar(p)
plt.show()
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contour(xx,yy,z, cmap='coolwarm')
fig.colorbar(p)
plt.show()
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z, cmap='coolwarm')
fig.colorbar(p)
plt.show()
```
