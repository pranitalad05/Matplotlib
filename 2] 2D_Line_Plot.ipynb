```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

#### 2D Line-Plot

```python
# 2D Line-Plot

# >> Bivariate Analysis
# >> Categorical_Numerical and Numerical-Numerical
# >> Use Case - Time series data
```

```python
#Plotting Simple Function

price=[48000,54000,57000,49000,47000,45000]
year=[2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
```

```python
#IPL Data Plotting

batsman = pd.read_csv(r"C:\Users\prani\Downloads\sharma-kohli (1).csv")
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#Multiple Plottings

plt.plot(batsman['index'],batsman['V Kohli']) #blue
plt.plot(batsman['index'],batsman['RG Sharma']) #orange
```

```python
#Title

plt.plot(batsman['index'],batsman['V Kohli']) #blue
plt.plot(batsman['index'],batsman['RG Sharma']) #orange

plt.title('Rohit Sharma VS Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#Colors(hex) 

plt.plot(batsman['index'],batsman['V Kohli'], color='#D9F10F') #Hex of color
plt.plot(batsman['index'],batsman['RG Sharma'], color='black') 

plt.title('Rohit Sharma VS Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#Linestyle - solid, dashed, dotted, dashdot

plt.plot(batsman['index'],batsman['V Kohli'], color='#D9F10F',linestyle='solid') 
plt.plot(batsman['index'],batsman['RG Sharma'], color='black',linestyle='dashdot') 

plt.title('Rohit Sharma VS Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#Linewidth

plt.plot(batsman['index'],batsman['V Kohli'], color='#D9F10F',linestyle='solid',linewidth=3) 
plt.plot(batsman['index'],batsman['RG Sharma'], color='black',linestyle='dashdot',linewidth=2) 

plt.title('Rohit Sharma VS Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#Marker(Style and Size) - . (dot), + (plus), D, >, <, O 

plt.plot(batsman['index'],batsman['V Kohli'],linestyle='solid',marker='D',markersize=5) 
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashdot',marker='D',markersize=4) 

plt.title('Rohit Sharma VS Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#Legend (called after using label method)
#loc - decides location for plotting of legend (best, upper right, upper left, lower left, 
                                                # lower right, right, center left, center right, 
                                                # lower center, upper center, center

plt.plot(batsman['index'],batsman['V Kohli'],marker='D',markersize=5,label='Virat') 
plt.plot(batsman['index'],batsman['RG Sharma'],marker='D',markersize=4,label='Rohit') 

plt.title('Rohit Sharma VS Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(loc='best') #Figures out the best location
```

```python
# Limiting Axes

price=[48000,54000,57000,49000,47000,45000,45000000]
year=[2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017.0,2019.0)
```

```python
#Grid

plt.plot(batsman['index'],batsman['V Kohli'],marker='D',markersize=5,label='Virat') 
plt.plot(batsman['index'],batsman['RG Sharma'],marker='D',markersize=4,label='Rohit') 

plt.title('Rohit Sharma VS Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
```

```python
#Show

plt.plot(batsman['index'],batsman['V Kohli'],marker='D',markersize=5,label='Virat') 
plt.plot(batsman['index'],batsman['RG Sharma'],marker='D',markersize=4,label='Rohit') 

plt.title('Rohit Sharma VS Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
plt.show() #to plot the graph
```
