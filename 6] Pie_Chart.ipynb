```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

#### Pie Chart

```python
# Univariate/Bivariate Analysis
# Categorical vs Numerical
# Use case - To find contribution on a standard scale
```

```python
# Simple Data

data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)
plt.show()
```

```python
# Dataset 

df = pd.read_csv(r'C:\Users\prani\Downloads\gayle-175.csv')
df
```

```python
# Percentage

plt.pie(df['batsman_runs'],labels=df['batsman'],autopct = '%0.1f%%')
plt.show()
```

```python
#Colors

plt.pie(df['batsman_runs'],labels=df['batsman'],autopct = '%0.1f%%',colors=['lightskyblue','lightgreen','lightyellow','lightpink','plum','violet'])
plt.show()
```

```python
# Explode and Shadow

plt.pie(df['batsman_runs'],labels=df['batsman'],autopct = '%0.1f%%',explode=[0.3,0,0,0,0,0],shadow=True)
plt.show()
```
