```python
# Side by side plotting of bar charts
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
df = pd.read_csv(r"C:\Users\prani\Downloads\batsman_season_record.csv")

plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'],width=0.2,color='lightgreen') #No spaces for color
plt.bar(np.arange(df.shape[0]), df['2016'],width=0.2,color='lightpink')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'],width=0.2,color='lightskyblue')

plt.xticks(np.arange(df.shape[0]),df['batsman'])

plt.show() #To avoid printing info at output and shows only bar chart
```

```python
# xticks(rotation) To avoid overlapping of titles on x-axis

children = [10,20,40,10,30]
colors = ['red red red','blue blue bluee','green green green','yellow yellow yellow', 'pink pink pinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Stacked bar Chart - bottom

plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=df['2016'] + df['2017'],label='2015')

plt.legend()
plt.show()
```
