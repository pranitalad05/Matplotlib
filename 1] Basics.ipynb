```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]

plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]

plt.plot(x,y, label="first", linewidth=4)
plt.plot(x2,y2, label="second", linewidth=4)
plt.legend()
plt.show()
```

```python
plt.scatter(x,y, label="first", linewidth=4)
plt.scatter(x2,y2, label="second", linewidth=4)
plt.legend()
plt.show()
```
